```python
#Import libraries
import pandas as pd
```

```python
# Load datasets
job_df = pd.read_csv("job.csv")
resume_df = pd.read_csv("resume.csv")
```

```python
#Remove null values and duplicates
job_df.drop_duplicates(inplace=True)
resume_df.drop_duplicates(inplace=True)
job_df.fillna("", inplace=True)
resume_df.fillna("", inplace=True)
```

```python
print(job_df.columns)
```

```python
job_df.columns.tolist()
```

```python
job_df["combined_text"] = (
    job_df["job_title"] + " " +
    job_df["job_description"] + " " +
    job_df["sector"] + " " +
    job_df["organization"]
)
```

```python
print(resume_df.columns)
```

```python
resume_df["combined_resume"] = resume_df["Resume_str"]
```

```python
print(job_df[["combined_text"]].head())

print(resume_df[["combined_resume"]].head())
```

```python
job_df.sample(1000, random_state=42).to_csv(
    "clean_job_small.csv",
    index=False
)

resume_df.sample(200, random_state=42).to_csv(
    "clean_resume_small.csv",
    index=False
)
```
